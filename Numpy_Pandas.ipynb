```python
#replacing everyone with 0 and changing data type of column to float
# changes are only for that cell if you dont assign that value to specific coloumn
df['Price'] = df['Price'].replace('Everyone','0')
df['Price']=df['Price'].str.replace('$','').astype(float)
```

```python
df['Size'].unique()
```

```python
df['Size']=df['Size'].str.replace('Varies with device','0').str.replace('k','').str.replace('M','').str.replace('+','').str.replace(',','').astype(float)
df['Size'].unique()
```

```python
df.drop_duplicates(subset=['App'])
```

```python
df=df.drop_duplicates(subset=['App'])
```

```python
df.shape
```

```python
# replacing null value from rating coloumn
df['Rating'].unique()
```

```python
df.Rating.dtype
```

```python
df['Rating'].mean()
```

```python
# if both have big difference then fill with median
df['Rating'].median()
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
df['Rating'].unique()
```

```python
df['Reviews']=df['Reviews'].str.replace('M','').astype(float)
```

```python
df['Reviews'].dtype
```

```python
df.head()
```

```python
# data normalisationas
# normalising reviews coloums
df['Reviews']=(df['Reviews']-df['Reviews'].min())/(df['Reviews'].max()-df['Reviews'].min())
df['Reviews'].unique()
```

```python
df['Popularity'] = df['Rating'] * df['Reviews']
```
